# Generate example time-domain signal (a combination of two sine waves)

```python
#!pip install sounddevice scipy matplotlib numpy
# !pip install librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
from scipy.io.wavfile import read
import sounddevice as sd
import librosa
import librosa.display
from scipy.io.wavfile import write


# Generate example time-domain signal (a combination of two sine waves)
fs = 1000  # Sampling frequency in Hz
t = np.linspace(0, 1, fs, endpoint=False)  # Time vector
f1, f2 = 5, 20  # Frequencies of the sine waves
signal = 2 * np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)

# Compute the frequency-domain representation using FFT
frequencies = np.fft.fftfreq(len(t), d=1/fs)
magnitude = np.abs(np.fft.fft(signal))

# Filter positive frequencies for visualization
positive_freq_indices = frequencies >= 0
frequencies = frequencies[positive_freq_indices]
magnitude = magnitude[positive_freq_indices]

# Plot the time-domain and frequency-domain visualizations
plt.figure(figsize=(12, 6), facecolor='#2F4F4F')

# Time-Domain Plot
plt.subplot(2, 1, 1, facecolor='#2F4F4F')
plt.plot(t, signal, color='white')
plt.title('Time-Domain Signal', color='white', fontsize=16)
plt.xlabel('Time (s)', color='white', fontsize=14)
plt.ylabel('Amplitude', color='white', fontsize=14)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.tick_params(colors='white', labelsize=12)

# Frequency-Domain Plot
plt.subplot(2, 1, 2, facecolor='#2F4F4F')
plt.stem(frequencies, magnitude, linefmt='white', markerfmt='o', basefmt="gray")
plt.title('Frequency-Domain Signal', color='white', fontsize=16)
plt.xlabel('Frequency (Hz)', color='white', fontsize=14)
plt.ylabel('Magnitude', color='white', fontsize=14)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.tick_params(colors='white', labelsize=12)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

# Record audio and plot the time-domain signal

```python

# Parameters for recording
sample_rate = 44100  # 44.1 kHz
duration = 5  # Duration of recording in seconds

# Record audio
print("Recording... Speak now!")
audio_data = sd.rec(int(sample_rate * duration), samplerate=sample_rate, channels=1, dtype='float64')
sd.wait()  # Wait for recording to finish
print("Recording complete.")

# Save audio as a WAV file
write("recorded_audio.wav", sample_rate, (audio_data * 32767).astype(np.int16))

# Normalize audio data for plotting
normalized_audio = audio_data.flatten() / np.max(np.abs(audio_data))

# Time axis for plotting
time_axis = np.linspace(0, duration, len(normalized_audio))

# Plot the recorded audio on a green background
plt.figure(figsize=(10, 4), facecolor='#2F4F4F')
plt.gca().set_facecolor('#2F4F4F')  # Set the background color of the plot area

# Plot the signal
plt.plot(time_axis, normalized_audio, color='white')
plt.title("Recorded Voice Signal", color='white', fontsize=16)
plt.xlabel("Time (s)", color='white', fontsize=14)
plt.ylabel("Amplitude", color='white', fontsize=14)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.tick_params(colors='white', labelsize=12)

# Adjust layout and display
plt.tight_layout()
plt.show()


```

# Perform the Short-Time Fourier Transform and compute the spectrogram

```python

# Load the recorded audio file
sample_rate, audio = read("recorded_audio.wav")

# If audio has more than one channel, use only the first channel
if len(audio.shape) > 1:
    audio = audio[:, 0]

# Normalize the audio
audio = audio / np.max(np.abs(audio))

# Define parameters for the Short-Time Fourier Transform
nperseg = 1024  # Length of each segment
noverlap = 512  # Overlap between segments

# Perform the Short-Time Fourier Transform and compute the spectrogram
frequencies, times, Sxx = spectrogram(audio, fs=sample_rate, nperseg=nperseg, noverlap=noverlap)

# Plot the spectrogram with a green background
plt.figure(figsize=(10, 6), facecolor='#2F4F4F')
ax = plt.gca()
ax.set_facecolor('#2F4F4F')  # Set the background color for the plot area

# Display the spectrogram
pcm = plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud', cmap='viridis')
plt.title("Spectrogram (Short-Time Fourier Transform)", color='white', fontsize=16)
plt.xlabel("Time (s)", color='white', fontsize=14)
plt.ylabel("Frequency (Hz)", color='white', fontsize=14)

# Add and customize the color bar
cbar = plt.colorbar(pcm)
cbar.set_label("Power (dB)", color='white', fontsize=14)
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(cbar.ax.yaxis.get_ticklabels(), color='white')

# Customize ticks and grid
plt.tick_params(colors='white', labelsize=12)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
# Mel-Frequency Cepstral Coefficients
```

```python

# Load the audio file
audio_file = "recorded_audio.wav"  # Replace with the path to your audio file
audio, sample_rate = librosa.load(audio_file, sr=None)

# Compute MFCCs
mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=13)

# Plot the MFCCs with a green background
plt.figure(figsize=(10, 6), facecolor='#2F4F4F')
ax = plt.gca()
ax.set_facecolor('#2F4F4F')  # Set the background color of the plot area

# Display MFCCs
img = librosa.display.specshow(mfccs, x_axis='time', sr=sample_rate, cmap='viridis')
cbar = plt.colorbar(img, format='%+2.0f dB')
cbar.set_label("Power (dB)", color='white', fontsize=14)
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(cbar.ax.yaxis.get_ticklabels(), color='white')

# Customize text, labels, and ticks
plt.title('Mel-Frequency Cepstral Coefficients (MFCCs)', color='white', fontsize=16)
plt.xlabel('Time (s)', color='white', fontsize=14)
plt.ylabel('MFCC Coefficients', color='white', fontsize=14)
plt.tick_params(colors='white', labelsize=12)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
